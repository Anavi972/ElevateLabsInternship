```python
import sqlite3
import pandas as pd
```

```python
conn= sqlite3.connect("ElevateLabs.db")
```

```python
df.to_sql("Superstore",conn,if_exists="replace",index=False)
```

```python
pd.read_sql("select * from Superstore limit 5",conn)
```

```python
query= """
select "sub-category", sum(sales) as totalsales, sum(quantity) as totalquantity
from superstore
group by
"sub-category"
"""
a=pd.read_sql(query,conn)
a
```

```python
query2= """
select "order id" 
from superstore
where region="South" and category="Furniture"
"""
pd.read_sql(query2,conn)
```
